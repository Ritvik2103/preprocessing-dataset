null values can be eliminated by either deleting the rows with null values or filling the null spaces with a fixed value
but eliminating a large number of rows causes improper analysis of the dataset

Thus we can implement the use of an imputer to obtain a null free dataset by filling it up with a fixed value

```python
from sklearn.impute import SimpleImputer
```

```python
impute= SimpleImputer(missing_values=np.NaN, strategy=('most_frequent'))  
#strategy refers to the const value that is to be filled
#we can fill the null values with the mean, median, mode or a const value
#here we impy the use of the mode strat
```

```python
impute.fit(df['Rating'])
```

the fit function takes a 2D array as an Input thus we have to reshape the 1D array to a 2D 


```python
lst= []
for i in df['Rating']:
    lst.append(i)
```

```python
arr= np.array(lst)    #converting the list to an array
```

```python
arr.shape    #thus the input array is a 1D array so we reshape it to 2D
```

```python
arr= np.reshape(arr,(10841,1))  #reshaping the array
 
```

```python
arr.shape
```

```python
#now we can use the impute fit func
```

```python
impute.fit(arr)   #this says the fitting is successfull and now we can implement the transform function
                                                                
```

```python
df['Rating']= impute.transform(arr)      #implementing the transform func
```

```python
df['Rating']      #we have now replaced the null values with the most frequently obtained rating
```

```python
df['Rating'].isnull().sum()     #thus no null values are present
```

### PREPROCESSING

we see that a few columns contain numbers are in the string fromat, for analysing such data we convert them to integers or floats(depending on the datatype)


STARTING WITH THE NUMBER OF INSTALLS:
instead of having a '+' at the end of each value we just replace it with the number that is free of ','

```python
df['Installs']
```

```python
install=[]
for i in df['Installs']:
    
    install.append(int(i[:-1].replace(',', '')))
    
```

```python
df['Installs']= install  
```

```python
for i in df['Installs']:
    print(i, type(i))
    break                               #thus the installs column is preprocessed
```

```python
df
```

PREPROCESSING PRICE OF APPS

```python
df['Price'].unique()
```

```python
## removal of the $ sign
```

```python
price=[]
for i in df['Price']:
    if('$' in i):
        price.append(float(i[1:]))
    else:
        price.append(float(i))
    
```

```python
set(price) # gives all the unique values present in the list
```

```python
df['Price']= price     #the price column is preprocessed
```

```python
df=df.rename(columns={'Price': 'Price in USD'})     #renaming the price column
```

```python
df
```

PREPROCESSING SIZE

```python
df['Size'].unique()
```

```python
#this column contains various formats: 'varies with device', sizes in 'M', sizes in 'k'
```

```python
size=[]
for i in df['Size']:
    if('M' in i):
        size.append(int(float(i[:-1])*1e3))       #converting M to k
    if('k' in i):
        size.append((float(i[:-1])))
    if('Varies' in i):
        size.append(int(10))      #let the size be 10 k
    
        
        
        
```

```python
df['Size']= size      #preprocessed size
```

```python
df= df.rename(columns={'Size': 'Size(k)'})    #renaming
```

```python
df
```
